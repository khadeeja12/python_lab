we can do visualization in python with this library.
inside pyplot , matplotlib exist. we dont create any datasets here.just use x and y. same number of values should put for x and y.

```python
import matplotlib.pyplot as plt
import numpy as np

#Line Plot
x=[1,2,3,4]
y=[11,22,33,44]

plt.plot(x,y,color='red',linestyle = '-', marker='o',label='Line Plot')
plt.legend() #topleft box is the legend
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

**Scatter plot**

```python
import matplotlib.pyplot as plt
import numpy as np

x=[8,2,6,3,5,9,1]
y=[9,4,2,6,7,1,8]

plt.scatter(x,y,color='blue',label='Scatter Plot',marker='*')
plt.legend()
plt.title('Scatter Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

**Bar plot**

```python
import matplotlib.pyplot as plt
import numpy as np

categories=['A','B','C','D']
values=[10,20,30,40]

#plt.bar(categories,values,color='green',label='Bar Plot') vertical
plt.barh(categories,values,color='green',label='Bar Plot') #horizontal
plt.legend()
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

**Histogram**

```python
import matplotlib.pyplot as plt

data=[4,2,7,2,7,4,6,3,9,6,6,6,8,5,9,9,2,2]

plt.hist(data,bins=5,color='orange',label='Histogram') # bins means no of values repeating.
plt.legend()
plt.title('Histogram')
plt.xlabel('Data')
plt.ylabel('Frequency')
plt.show()
```

**piechart**

```python
labels=['Apple','mango','orange','banana']
sizes=[135,215,245,300]
colors=['gold','yellowgreen','lightcoral','lightskyblue']

explode=(0.1,0,0,0)
plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True,startangle=90,explode=explode)
plt.axis('equal')
plt.show()
```

**Box plot**

```python
import matplotlib.pyplot as plt
import numpy as np

data=[[2,3,5,8,12],[1,4,7,10,14],[0,2,6,8,13]]

plt.boxplot(data,vert=True,patch_artist=True,labels=['A','B','C'])
plt.title('Box Plot')
plt.ylabel('Values')
plt.show()
```

**Viloin Plot** it says the concentration

```python
import matplotlib.pyplot as plt
import numpy as np

data=[[2,3,5,8,12],[1,4,7,10,14],[0,2,6,8,13]]

plt.violinplot(data,showmeans=True,showmedians=True)
plt.title('Viloin Plot')
plt.ylabel('Values')
plt.show()
```
