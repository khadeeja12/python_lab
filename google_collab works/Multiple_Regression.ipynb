**Multiple Regression**

```python
import pandas
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler  #it is needed for scaling data

df = pandas.read_csv("data_scale.csv")
df.head()
df.info()
print(df)

X = df[['Weight', 'Volume']]
y = df['CO2']

scale=StandardScaler()
scaledX=scale.fit_transform(X)

regr = linear_model.LinearRegression()
regr.fit(scaledX, y)
```

```python
predictedCO2 = regr.predict([[2300,1300]])

print(predictedCO2)
print(regr.coef_)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

x=np.random.normal(32,5,100)
y=np.random.normal(69,8,100)/x

plt.scatter(x,y)
plt.show()
```

```python
# Splitting the data into training and testing sets
train_x = x[:80]
train_y = y[:80]
test_x = x[80:]
test_y = y[80:]

# Plotting the training data
plt.scatter(train_x, train_y)
plt.xlabel('train_x')
plt.ylabel('train_y')
plt.title('Scatter Plot of Training Data')
plt.show()
```

```python
mymodel = np.poly1d(np.polyfit(train_x, train_y, 4)) # we draw a line using polynomial regression
myline = np.linspace(0, 100, 100)

plt.scatter(train_x, train_y)
plt.plot(myline, mymodel(myline))
plt.show()
```

```python
from sklearn.metrics import r2_score

# Assuming you have a trained model, e.g., 'model' and its predictions
# Replace 'mymodel' with your actual model's prediction function

# Calculate R² score for the training set
train_predictions = mymodel(train_x)  # Replace 'mymodel' with your actual model
train_r2 = r2_score(train_y, train_predictions)
print("Train:", train_r2)

# Calculate R² score for the test set
test_predictions = mymodel(test_x)  # Replace 'mymodel' with your actual model
test_r2 = r2_score(test_y, test_predictions)
print("Test :", test_r2)
```
